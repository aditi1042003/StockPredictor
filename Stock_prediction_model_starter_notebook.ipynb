```python
#importing required libraries
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
# Set the start and end date
# Here is 12 years of data , can be formated as per need of model
start_date = '2010-01-01'
end_date = '2022-01-01'

# Set the ticker i.e selecting a stock by its stock symbol
#for more info visit: https://finance.yahoo.com/
# example for Applle Inc: ticker='AAPL' , for Google ticker='GOOGL'
ticker = 'GOOGL'

# Get the data
data = yf.download(ticker, start_date, end_date)

# Print the last 5 rows
data.tail()
```

```python
# adding index to data set
data=data.reset_index()
data.head()

```

```python
#removing not needed columns
data=data.drop(['Adj Close'],axis=1)
data.head()
```

```python
#removing not needed columns
data=data.drop(['Date'],axis=1)
data.head()
```
